```python
# this exercise is about dictionaries in python
# Dictionaries in python is one of the collection datatype used to store multiple items and and items can be of any data type and it store the item in form of keys and values.
```

```python
import time
last_login = time.time()
```

```python
# creation of dictionary

user = {
    "user_name" : "devkantkumar",
    "role" : "admin",
    "__id" : "a08765esxcvbnmki765",
    "access_type" :"full_access",
    "is_on_subscription" : True,
    "last_login" : last_login
}

print("length of dictionary : " + str(len(user)))
```

```python
# accessing the items of dictionary

print(f'user name : {user.get("user_name")} has role of : {user["role"]} and last login at : {user["last_login"]}')
```

```python
#accessing the keys of dictionaries
print(f'user have these properties : ' {user.})
```
