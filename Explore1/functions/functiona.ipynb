Function in python : 
 - Function is a block of code that are written to preform one specific task
 - Its is defined using def keyword
 - It reduces code redundancies
 - Helps to implement code reusability
 - It can take input and return values and these value can be of any data type like (string,int,float,list,tuple,set,dictionary,etc)
 - Input values to function is called argument and passed during function call
 - Parameter are used to collect the passed value to function 

```python
user = {
    "username" : "devkantkumar",
    "first_name" : "Dev",
    "middle_name" : "Kant",
    "last_name" : "Kumar",
    "is_active" : True,
    "notifications" : 100,
    "posts" : [
        {
            "title" : "Functions in python",
            "description" : """ Complete function in python | From Basic to pro level
             topics covered :
              - what is function
              - why function
              - parameters and arguments
              - scope of function : global , local and nonlocal
              - default parameters
              - positional and keyword arguments
              - *args vs **kwargs
              - decorators
              - lambda function
              - recursion
              - generator
                """,
            "tags" : ["python","python-programming","functions-python","programming"]
        }
    ]
}
print(f"user info : {user}")
```

```python
def get_user_fullname(user):
    f_name = user.get("first_name") + " "
    m_name = user.get("middle_name") + " "
    l_name = user.get("last_name")
    return f_name + m_name + l_name

def get_user_post(user):
    posts = user.get("posts")
    return posts
```

```python
full_name = get_user_fullname(user)
print(f"Full name : {full_name}")
```

```python
# function with none return value : function will return none if return statement is not used
def show_user_posts(user):
    user_posts = get_user_post(user)
    for x , y in user_posts[0].items():
        print(f"{x} : {y}")
```

```python
show_user_posts(user)
```

```python
x = 100 #global scope . it can be accessed from everywhere in this program

print("x is " + str(x))

def change_value(new_val):
    current_step = 1 #local scope . can only access inside this function
    global x
    x = new_val
    print("Current step : " + str(current_step))

change_value(550)
print("x is now : " + str(x))
# print("current_step is : " + str(current_step)) not accessible outside the function , its local to change_value function
```

Default parameter : its the optional parameters and if its value is provided in function call the provided value will be used and if not default value will be used

```python
def make_payment(receiver,amount,note = "Money Transfer"):
    print(f"Payment order created")
    print(f"Sending {amount} to {receiver} .........")
    print(f"Transaction successful with note : {note}")
```

```python
make_payment("Dev",500000)
```

Positional Arguments vs Keyword Arguments
 - Positional Arguments : Only values in passed to function and order matter most
 - Keyword Arguments : Values is passed as key : value pair and order doesn't matter

```python
make_payment("Adam",100000)
print("=" * 10)
make_payment(receiver = "Bob",amount= 500000,note = "Business Payment")
```

Only positional arguments vs Only keyword arguments
  - only positional arguments : use ,/ at the end of the parameter list
  - only keyword arguments : use *, at the start of the parameter list

```python
def sum(a,b,c,d,/):
    return a + b + c + d
```

```python
loan_processing_charge = 1000
rate_of_interest = 3

def process_loan(*,lender,amount,time):
    print(f"Hi {lender}, Your loan enquiry is received")
    principal_amount = (amount * rate_of_interest *  time )/100
    print(f"Total Amount to repay will be : {principal_amount + amount + loan_processing_charge}")
    print(f"Loan processing charge : {loan_processing_charge}")
    print(f"Rate of Interest is : {rate_of_interest}% per annum")
    print(f"Within the span of {time} years")
```

```python
#for process loan function we need to send the args with key because incorrect order of args with leads to error .
process_loan(lender = "Ava",amount = 1000,time = 2)
```

*args vs **kwargs
  - *args : arbitrary positional argument is used to collect all passed args to function when the no of args pass is unknown and these are stored in tuple as function parameters
  - **args : arbitrary keyword argument is used to collect all args to function when the no of args is unknown  (args = keyword args) and these are stored in dictionary as function parameters

```python
def sum_of_all(*args):
    total = 0
    print(f"type of args : {type(args)} \nargs are : {args} \nargs length : {len(args)}")
    for num in args:
        total += num
    return total
```

```python
print(f"sum is : {sum_of_all(1, 2, 3, 4, 5, 6, 7, 89)}")
print("=" * 10)
print(f"sum is : {sum_of_all(1, 2, 3, 4, 5)}")
```

```python
def new_user(**args):
    print(f"type of args : {type(args)} \nargs are : {args} \nargs length : {len(args)}")
    for key , value in args.items():
        print(f"{key} : {value}")
```
