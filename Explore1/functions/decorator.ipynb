Decorators in python is a function which can take another function as argument as return another function

```python
def decorator(fun):
    def wrapper(*args,**kwargs):
        print("Transaction initiated")
        fun(*args,**kwargs)
        print("Transaction successfull")
    return wrapper
```

```python
@decorator
def process_payment(sender,receiver,amount,note = "money transfer"):
    print(f"{sender} sending ${amount} to {receiver} , note : {note}")
    print(f"{receiver} received ${amount} from {sender},note : {note} ")

# def process_payment(*args,**kwargs):
#     print(args)
#     print(kwargs)
```

```python
process_payment("Jhon","Bob",50000,note = "Business Payment")
```

Use case of decorator
- logging : track when functions are called
- timing : measure execution time
- authentication : check permissions before running
- caching :  store results to avoid recalculation
- validation : check inputs before processing
