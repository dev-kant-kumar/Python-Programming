Recursion :
  - When a function call itself to perform some task repeatedly.
  - Its has base condition where the recursion stop and return the value to previous layer in stack .
  - In case of no base condition , recursion run infinite times and end up overflowing the stack .
  - Python stop it with message : RecursionError: maximum recursion depth exceeded in case of no base condition

we can check the depth limit of recursion and also modify it if we need more depth but do with caution it might crash the program


```python
import sys
print(sys.getrecursionlimit())

sys.setrecursionlimit(4000)
print(sys.getrecursionlimit())
```

```python
# a recursive function to print the number from n to 1
def print_num(n):
    if n == 0 :
        return
    print(n ,end = " ")
    print_num(n-1)
```

```python
print_num(10)
```

```python
# find sum of n natural number using recursion
def sum_of_nums(n):
    if n == 1 :
        return 1
    return sum_of_nums(n - 1) + n
```

```python
 print(sum_of_nums(50))
```

```python
# find the factorial of n using recursion

def fact(n):
    if n == 1 :
        return 1
    return fact(n-1) * n
```

```python
print(fact(5))
```

```python
# print the elements of list using recursion
def print_list(list_data,idx = 0):
    if idx == len(list_data):
        return
    print(list_data[idx],end = " ")
    print_list(list_data,idx+1)
```

```python
papers = ["networking","operating system","database","dsa"]
print_list(papers)
```

```python
marks = [98,99,97,99,98]
print_list(marks)
```
